# Thresholding
Thresholding is a type of image segmentation, where we change the pixels of an image to make the image easier to analyze. In thresholding, we convert an image from color or grayscale into a binary image.Most frequently, we use thresholding as a way to select areas of interest of an image, while ignoring the parts we are not concerned with.

# 1- import libraries

```python
import imageio.v3 as iio
import numpy as np
import skimage.color
import skimage.util
import matplotlib.pyplot as plt
%matplotlib widget
```

we used a simple NumPy array manipulation to separate the pixels belonging to the root system of a plant from the black background.

# 2- load the original image

```python
# read the MRI image as grayscale from the outset
image = iio.imread(uri="./irm.jpg", mode="L")
```

```python
fig, ax = plt.subplots()
plt.imshow(image, cmap="gray")
```

# 3- Convert the image to grayscale

```python
# convert the image to float dtype with a value range from 0 to 1
image = skimage.util.img_as_float(image)
```

# 4- denoise the image
When we blur an image, we make the color transition from one side of an edge in the image to another smooth rather than sudden. The effect is to average out rapid changes in pixel intensity.
We used "skimage.filters" module and we used the Gaussian blur.

```python
# blur the image to denoise
blurred_image = skimage.filters.gaussian(image, sigma=1.0)
```

# 5- Display the denoised image

```python
fig, ax = plt.subplots()
plt.imshow(blurred_image, cmap="gray")
```

# 6- Create the histogram

```python
histogram, bin_edges = np.histogram(image, bins=256, range=(0, 1))
```

```python
# configure and draw the histogram figure
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.xlim([0.0, 1.0])  # <- named arguments do not work here

plt.plot(bin_edges[0:-1], histogram)  # <- or here
```

# 3- create a mask based on the threshold
Since the image has a black background, most of the pixels in the image are black. This corresponds to what we see in the histogram: there is a peak near the value of 0.0.

If we want to select the white ubstance and the grey, le liquide cphalorachidien. and not the background, we want to turn off the black background pixels, while leaving the pixels for the white and the grey substant turned on. So, we should choose a value of t somewhere in the middle and turn pixels above that value off. we tried different thresholds and we chose t=0.5

```python
t = 0.6


binary_mask = blurred_image < t

fig, ax = plt.subplots()
plt.imshow(binary_mask, cmap="gray")
```
